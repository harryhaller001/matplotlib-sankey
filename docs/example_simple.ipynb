# Simple sankey plot

The function `sankey_simple` required as data input a numpy array of 2 dimensions.

```python
from matplotlib_sankey import sankey_simple
import numpy as np

data = np.array([[10, 24, 50, 80], [14, 30, 24, 40], [4, 10, 50, 53]])
```

```python
sankey_simple(data)
```

Use `max_scale=True` to allow indication of absolute difference between columns.

```python
sankey_simple(data, max_scale=True)
```

### Change apparance of sankey plot

* Choose colormap or custom colors
* Add spacing between ribbons with `spacing`.
* change ratio of column and ribbon width with arguments `column_width` and `ribbon_width`.
* Change transparency of ribbons with `alpha`.

```python
sankey_simple(
    data,
    title="Sankey plot",
    spacing=0.02,
    colors="coolwarm",
    column_width=2,
    ribbon_width=4,
    alpha=0.2,
)
```

Change type of curve.

```python
sankey_simple(data, curve_type="line", frameon=False)
```
